For this assignment, I want to see whether life expectancy have any relations with race and sex.

I found this dataset from [data.gov](https://catalog.data.gov/dataset/age-adjusted-death-rates-and-life-expectancy-at-birth-all-races-both-sexes-united-sta-1900) which highlights the death rates and life expectancy at birth by race and sex since 1900. However, I am only interested in trends from the past ten years (2005-2015).

---



```
import pandas as pd
```

```
life_expectancy = pd.read_csv('https://data.cdc.gov/api/views/w9j2-ggv5/rows.csv?accessType=DOWNLOAD')
life_expectancy
```

```
life_expectancy.columns
```

To make the dataset more readable, I renamed the columns. 

```
life_expectancy.columns = ['Year', 'Race', 'Sex', 'Avg Life Expectancy', 'Death Rate']
life_expectancy.columns
```

```
life_expectancy
```

```
life_expectancy.isnull()
```

```
life_expectancy.isnull().sum()
```

I want to start first with race. I want to see if race have any connections with life expectancy. As I mentioned earlier, I will only look at trends from 2005 and 2015. The Avg Life Expectancy value for 2015 are missing. I will replace the missing value with mean. However, in order to do so, I need to create a separate datasets for white and black people. After we calculated the mean for both, I can combine the two datasets.

```
white_exp = life_expectancy[life_expectancy.Year >= 2005] [life_expectancy.Sex == 'Both Sexes'] [life_expectancy.Race == 'White']
```

```
white_exp
```

Now let's the calculate the mean of the average life expectancy for white people.

```
mean1 = white_exp['Avg Life Expectancy'].mean()
mean1
```

I can now replace missing value with the mean.

```
white_exp['Avg Life Expectancy'].fillna(mean1, inplace = True)
white_exp
```

```
black_exp = life_expectancy[life_expectancy.Year >= 2005] [life_expectancy.Sex == 'Both Sexes'] [life_expectancy.Race == 'Black']
```

```
black_exp
```

This the mean for the avaerage life expectancy for black people from 2005 to 2015.

```
mean2 = black_exp['Avg Life Expectancy'].mean()
mean2
```

Now I replace the NA value with this mean.

```
black_exp['Avg Life Expectancy'].fillna(mean2, inplace = True)
black_exp
```

```
white_exp, black_exp
```

After calculating the mean, I can now combine the two datasets. We can even look at the table to make some comparison.

```
race_exp =pd.concat([white_exp,black_exp], ignore_index=True)
```

```
race_exp
```

However, it will be better to create a plot in order to visualize the trend. The data reveals that white people live slighly a longer life than black people.  Nevertheless,  the difference between the two races are small. I cannot rule out that there is a relation between race and life expectancy.

```
race_exp.groupby(['Year','Race'])['Avg Life Expectancy'].mean().unstack('Race').plot(kind='bar')
```

Let see if sex have any connection with life expectancy. The process is very similar to that of race. We have to do some data cleaning and perform some calculations (in this case we have to calculate the mean)

```
male_exp = life_expectancy[life_expectancy.Year >= 2005] [life_expectancy.Sex == 'Male'] [life_expectancy.Race == 'All Races']
```

```
male_exp
```

```
mean3 = male_exp['Avg Life Expectancy'].mean()
mean3
```

```
male_exp['Avg Life Expectancy'].fillna(mean3, inplace = True)
male_exp
```

```
female_exp = life_expectancy[life_expectancy.Year >= 2005] [life_expectancy.Sex == 'Female'] [life_expectancy.Race == 'All Races']
```

```
female_exp
```

```
mean4 = female_exp['Avg Life Expectancy'].mean()
mean4
```

```
female_exp['Avg Life Expectancy'].fillna(mean4, inplace = True)
female_exp
```

```
sex_exp =pd.concat([male_exp,female_exp], ignore_index=True)
```

```
sex_exp
```

Looking at the graph below, we can tell that women lives longer life than men. Male's life expectancy is 75 while female can live up to 80 years old. 

```
sex_exp.groupby(['Year','Sex'])['Avg Life Expectancy'].mean().unstack('Sex').plot(kind='bar')
```
